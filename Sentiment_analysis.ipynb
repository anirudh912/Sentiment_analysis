### Sentiment Analysis using Python NLTK 

```python
# Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer 
from tqdm.notebook import tqdm
plt.style.use('ggplot')
sia = SentimentIntensityAnalyzer ()
```

```python
# Upload data

df = pd.read_csv(r'C:\Users\91988\Downloads\Tweets.csv')
print(df.shape)
```

```python
df.head()
```

```python
# Vader sentiment scores

#example
sia.polarity_scores('I am sad now')
```

```python
sia.polarity_scores('I can go to the show now')
```

```python
# Run it on the entire dataset

scrs = {}
for i, row in tqdm(df.iterrows(), total = len(df)):
    text = str(row['text'])
    finid = row['textID']
    scrs[finid] = sia.polarity_scores(text)
```

```python
scrs
```

```python
vdr = pd.DataFrame(scrs).T
```

```python
vdr
```

```python
mood = []
for i in vdr.index:
    
    if vdr['compound'][i] > 0.0000:
        mood.append('Positive')
    elif vdr['compound'][i] < 0.0000:
        mood.append('Negative')
    else:
        mood.append('Neutral')
vdr.insert(len(df.columns), 'mood', mood, True)
print(vdr)
```
